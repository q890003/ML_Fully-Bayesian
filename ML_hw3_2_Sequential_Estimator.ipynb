```
import numpy as np
# import math
import random
# mean, var = float(input("mean= ")) ,float(input("var= ")) 
mean,var = 20, 5   #for test
sigma = var**(1/2)
print("Data point source function: N({mean}, {variance}) \n\n".format(mean = mean, variance = var))

old_mean, old_var = 0.0, 0.0
num = 0
# pi = math.pi
while True:
    num += 1
#generating data ~ N(mean, var).
    ##Boxâ€“Muller method to generate value under given N(0, 1), standard normal
    # U = np.random.uniform(0,1)
    # V = np.random.uniform(0,1)
    # z = (-2*math.log(U))**(-1/2) * math.cos(2*pi*V)
    ##Method2
    z = np.sum(np.random.uniform(0.0, 1.0, 12)) - 6

    #transform from standard normal back to N(mean, var).
    #formula: z = (x-m)/sigma -> x= z*sigma + mean
    gen_value = z *sigma + mean
    print("Add data point: {new_data}".format(new_data = gen_value))

#online algorithm
    new_mean = old_mean +  (gen_value-old_mean)/num
    new_var = old_var + ((gen_value-old_mean)*(gen_value-new_mean) - old_var) /num
    
    print("Mean = {new_mean}    Variance = {new_var}".format(new_mean = new_mean, new_var = new_var))
    
#Update
    if (abs(new_mean - old_mean) < 0.0001) and (abs(new_var - old_var) < 0.0001):
        break
    old_mean , old_var = new_mean, new_var
    
```
